# **Problem Statement**



With the rapid growth of cloud computing, artificial intelligence (AI), and digital services, data centers have become one of the largest consumers of electricity and water worldwide. Managing these resources efficiently while minimizing environmental impact has become a significant challenge for organizations. Factors such as cooling systems, Power Usage Effectiveness (PUE), Water Usage Effectiveness (WUE), facility capacity, and regional water stress directly influence operational efficiency and sustainability. An in-depth exploratory analysis is required to identify consumption patterns, evaluate resource utilization, and uncover opportunities for improving efficiency and supporting sustainable data center operations.

#### **Define Your Business Objective?**

The primary objective of this project is to perform an Exploratory Data Analysis (EDA) on global data center operational data to understand electricity and water consumption patterns, assess the impact of cooling systems and infrastructure efficiency, compare resource usage across facility types and geographical locations, and identify key factors affecting sustainability. The insights generated from this analysis can help organizations optimize resource utilization, reduce operational costs, improve energy and water efficiency, and support environmentally responsible decision-making.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Create a copy of the original dataset
df1 = df.copy()

# Check for missing values
print("Missing Values:")
print(df1.isnull().sum())

# Check for duplicate rows
print("\nDuplicate Rows:", df1.duplicated().sum())

# Remove duplicate rows
df1.drop_duplicates(inplace=True)

# Check data types
print("\nData Types:")
print(df1.dtypes)

# Check dataset shape after cleaning
print("\nDataset Shape:", df1.shape)

# Reset index
df1.reset_index(drop=True, inplace=True)

print("\nDataset is ready for analysis.")
````
