#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code

# Correlation Heatmap of Numerical Variables

plt.figure(figsize=(10,7))

numerical_cols = ['Estimated_Capacity_MW',
                  'PUE',
                  'WUE_L_per_kWh',
                  'Daily_Electricity_Usage_MWh',
                  'Daily_Water_Usage_Gallons']

sns.heatmap(df1[numerical_cols].corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f',
            linewidths=0.5)

plt.title('Correlation Heatmap of Numerical Variables')

plt.show()
````

##### 1. Why did you pick the specific chart?

A Heatmap is the best visualization for understanding the correlation among multiple numerical variables simultaneously. It quickly highlights strong positive, negative, or weak relation

##### 2. What is/are the insight(s) found from the chart?

It helps identify highly correlated variables that may influence each other.
Weak correlations indicate variables that are relatively independent.
